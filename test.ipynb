# 对比学习

```python
from preprocess import prepare_for_cl
import pandas as pd
events = pd.read_csv('event_extraction.csv')
price = pd.read_csv('carbon_price.csv')
data = prepare_for_cl(price, events, text=['type'])
```

```python
import torch
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import torch.nn as nn
from constrative import SupConNet, SupConLoss

input_data_array = np.vstack(data['embedding'].values)
output_data = data['reward'].values

features = torch.tensor(input_data_array, dtype=torch.float32)
labels = torch.tensor(output_data, dtype=torch.long)
dataset = TensorDataset(features, labels)

batch_size = 32
train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

siamese_net = SupConNet(input_dim=768)
optimizer = optim.SGD(siamese_net.parameters(), lr=0.001)
loss_function = SupConLoss(contrast_mode='all')

num_epochs = 25
for epoch in range(num_epochs):
    for batch_idx, (features, labels) in enumerate(train_loader):
        optimizer.zero_grad()
        output = siamese_net(features)
        loss = loss_function(output.unsqueeze(1), labels)
        loss.backward()
        optimizer.step()

        if batch_idx % 10 == 0:
            print(f'Epoch [{epoch + 1}/{num_epochs}], Batch [{batch_idx + 1}/{len(train_loader)}], Loss: {loss.item()}')

features = []
labels = []
with torch.no_grad():
    for input_data, batch_labels in train_loader:
        output = siamese_net(input_data)
        features.extend(output.numpy())
        labels.extend(batch_labels.numpy())
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
embedded = tsne.fit_transform(features)
plt.figure(figsize=(10, 8))
plt.scatter(embedded[:, 0], embedded[:, 1], c=labels, cmap='viridis', alpha=0.8)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('t-SNE Visualization of Sample Features')
plt.colorbar(label='Class Label')
plt.show()
```
